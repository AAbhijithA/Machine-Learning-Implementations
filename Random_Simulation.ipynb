Importing Necessary Libraries

```python
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

Generating random 1500 data points

```python
np.random.seed(21)
data = np.random.rand(1500,2)
data
```

Running 1500 simulations for the line of best fit with scoring via R Square method

```python
x = data[:,0]
y = data[:,1]
simulation = 0
best_line = None
best_score = 0
while(simulation < 1500):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=500,random_state=simulation,shuffle=True)
    x_train = x_train.reshape(-1,1)
    x_test = x_test.reshape(-1,1)
    local_Line = linear_model.LinearRegression()
    local_Line.fit(x_train,y_train)
    local_score = local_Line.score(x_test,y_test)
    if(local_score > best_score):
        best_score = local_score
        best_line = local_Line
    simulation = simulation + 1
print(best_score)
```

Plotting the final result of the saved linear regression model

```python
plt.figure()
plt.scatter(x,y,s=1,c='blue')
plt.plot(x,best_line.predict(x.reshape(-1,1)),color='violet')
plt.legend(['Data Points','Line of Best Fit'])
plt.ylabel("Y")
plt.xlabel("X")
```
