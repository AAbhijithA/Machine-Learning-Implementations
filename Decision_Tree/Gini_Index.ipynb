Reading the data from the CSV file

```python
df = pd.read_csv('classification.csv')
df.head(10)
```

Function for Gini Index

```python
def gini(df,column,split_check,target):
    target_class = df[target].unique()
    left_child = df[df[column] <= split_check]
    right_child = df[df[column] > split_check]
    N = len(df)
    L = len(left_child)
    R = len(right_child)
    GiniL = 1
    GiniR = 1
    for t in target_class:
        target_class_left = left_child[left_child[target] == t]
        target_class_right = right_child[right_child[target] == t]
        if L > 0:
          pTarget_left = len(target_class_left) / L
          GiniL -= (pTarget_left**2)
        if R > 0:
          pTarget_right = len(target_class_right) / R
          GiniR -= (pTarget_right**2)
    Gini = ((L / N)*GiniL) + ((R / N)*GiniR)
    return Gini

def giniIndex_Feature(df,column,i,target):
    df = df.sort_values(by=column,ascending=True)
    best_Gini = 2
    best_split = None
    for j in range(0,len(df)):
        split_check = df.iloc[j,i]
        G = gini(df,column,split_check,target)
        if G < best_Gini:
            best_Gini = G
            best_split = split_check
    return best_Gini, best_split
```

Printing Gini Index for each column

```python
list_columns = list(df.columns)
for i in range(0,len(list_columns)):
    if list_columns[i] == "Purchased":
        continue
    Gini, split = giniIndex_Feature(df,list_columns[i],i,"Purchased")
    print("Gini Index for " + str(list_columns[i]) + " is: " + str(Gini) + ", With a split at: " + str(split))
```
